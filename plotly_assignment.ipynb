```python
import plotly.express as px
import seaborn as sns
```

```python
#Q1. Load the "titanic" dataset using the load_dataset function of seaborn.
    #Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

#Ans--
titan=sns.load_dataset("titanic")

px.scatter(x=titan.age,y=titan.fare)
```

```python
#Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
#Ans--
tips=px.data.tips()
px.box(tips,x="day",y="total_bill")
```

```python
#Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. 
#Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

#ANS--
px.histogram(tips,x= "sex",y="total_bill" ,pattern_shape="smoker",color="day")
```

```python
#Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
iris=px.data.iris()
px.scatter_matrix(iris,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"]
                  ,color="species")
```

```python
#Q5. What is Distplot? Using Plotly express, plot a distplot.
'''ANS--Distplot is a seaborn library function that is used to visualize a univariate distribution of observations.
It combines a histogram with a kernel density estimate, allowing to visualize the distribution of a variable along with its density curve.'''

px.histogram(tips,x="total_bill")
```
